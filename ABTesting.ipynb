<h2>4.1 Feature Matrix ‚Äî Echoes of Tomorrow vs Afterlight</h2>

**Purpose:** Establish a side-by-side comparison before the persona tests.

| Dimension | Echoes of Tomorrow | Afterlight |
|---|---|---|
| **Emotional Core** | Speculative empathy; possibility; branching futures | Reassurance; tenderness; courage to begin |
| **Narrative Structure** | Multiverse / multiple timelines; cinematic | Linear, intimate, dialogue-first |
| **Visual Identity** | Futuristic motifs, timeline arcs | Haloed circle; light‚Äìshadow gradient; serif+sans pairing |
| **Innovation Framing** | Tech as mirror + growth engine | Tech as lantern / companion across time |
| **Onboarding / Accessibility** | Potentially abstract; needs framing | Immediately human; easy to grasp |
| **Campaign Extensions** | AR booth; multi-channel storytelling; immersive modes | Mirror billboards; AR ‚Äúcall your future self‚Äù; #MyAfterlight |
| **Risk / Tradeoffs** | Can feel lofty or conceptual | Can feel sentimental / less ‚Äúinnovation-forward‚Äù |

> **Working hypothesis:** Echoes excels in *vision & differentiation*; Afterlight excels in *immediacy & adoption*. The optimal merge balances **imagination** (A) with **reassurance** (B).
</div>


```python
# 4.2 LLM Feature-Group Ratings (mirrors the A/B notebook style: two system messages, one call)

feature_groups = [
    "Emotional Core",
    "Narrative Structure",
    "Visual Identity",
    "Innovation Framing",
    "Onboarding & Accessibility",
    "Campaign Extensions"
]

rating_prompt = f"""
You will rate two products across feature groups on a 0-10 scale (0 = never buy / irrelevant, 10 = buy immediately / highly compelling).
Show just your ratings per group for Product A (Echoes of Tomorrow) and Product B (Afterlight), state which product you prefer overall, then briefly explain why.

Feature Groups:
{', '.join(feature_groups)}

Product A (Echoes of Tomorrow) ‚Äî Key cues:
- Speculative empathy; branching timelines; futuristic identity
- Visionary tone; immersive AR/demo possibilities; cinematic language

Product B (Afterlight) ‚Äî Key cues:
- Reassurance; intimate reflection; time & light metaphor
- Immediate human resonance; mirrors/billboards/AR call; dialogue-first
"""

messages = [
    {"role": "system", "content": "You are a consumer insights LLM focused on marketing resonance."},
    {"role": "system", "content": rating_prompt}
]

fg_response = query_LLM(messages=messages)
print(fg_response, "\n")
```

<h2>4.3 Segment Sensitivity Map</h2>

**Purpose:** Anticipate which audiences are more responsive to each product‚Äôs feature profile.

- **Early Adopters / Futurists / Creators** ‚Üí Likely to favor **Echoes of Tomorrow** for its speculative depth and novelty signals.  
- **Mainstream / First-time Users / High-stakes Deciders** ‚Üí Likely to favor **Afterlight** for its clarity, reassurance, and human anchoring.  
- **Therapists / Coaches / Educators** ‚Üí May appreciate **Afterlight**‚Äôs tenderness *and* **Echoes**‚Äô growth language; potential hybrid fit.  
- **Brand / Campaign Planners** ‚Üí Echoes = differentiation play; Afterlight = broad adoption play; **merge** = *differentiated accessibility*.

> **Implication:** Test creative that **begins** in Afterlight‚Äôs voice (clarity & safety) and **expands** into Echoes‚Äô promise (growth & possibility).
</div>


```python
# 4.4 Persona loop (same structure as the original notebook: persona system + rating system)

personas = [
    "college student deciding on a scholarship application",
    "mid-career professional considering a role change",
    "new parent managing uncertainty",
    "creator seeking momentum on a personal project"
]

persona_prompt_template = """
Read the summaries for two products:

Product A: Echoes of Tomorrow
- Speculative empathy, branching futures, cinematic tone, innovation-forward.

Product B: Afterlight
- Intimate reassurance, light‚Äìshadow metaphor, dialogue-first, immediately human.

As this consumer: {persona}
Rate A and B 0‚Äì10, show which you prefer, and briefly explain your decision with 2‚Äì3 reasons.
"""

for p in personas:
    msgs = [
        {"role": "system", "content": f"You are a consumer who is a {p}"},
        {"role": "system", "content": persona_prompt_template.format(persona=p)}
    ]
    out = query_LLM(messages=msgs)
    print(f"\n-----\nPersona: {p}\n{out}\n")
```
